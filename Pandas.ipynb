### What Is A Series?
- Series is 1 dimensional in nature such as an array. Series is a mutable data object. It means it can be updated, new items can be added and existing items can be deleted from the collection.
- It can hold data of any type.
- It can be instantiated with an array or a dictionary. The keys of the dictionary are used to represent indexes.
- The constructor of Series can be called by passing in data, index, data type and a Boolean flag indicating if we want to copy the data. If we don’t specify the index names then the indexes are labeled as integers starting from 0.
- Think of Series as Vertical Columns that can hold multiple rows.

```python
series_1 = pd.Series(data=[111, 222, 3], index = ['one','two','three'])
print(series_1)
print('-'*35)
series_2 = pd.Series([111, 222, 3])
print(series_2)
print('-'*35)

# accessing series
print('Acces via 0:', series_1[0])
print('Acces via "one":', series_1['one'])
print('-'*35)
print('Acces via 0:', series_2[0])
```

### What Is A Data Frame?
- Possibly the most used data structure in a data science project.
- It is a table with rows and columns — like a SQL Table Or Excel Spreadsheet Sheet.
- The table can store in memory data objects in different format.
- Offers high performing time series data analysis and engineering
- A data frame is a container of one or more Series.
- DataFrame is mutable.

```python
d1 = {'ColumnA' : pd.Series([111, 222, 333]),
     'ColumnB' : pd.Series([444, 555, 666])}

d2 = {'ColumnA' : [111, 222, 333],
     'ColumnB' : [444, 555, 666]}

df1 = pd.DataFrame(d1)
df2 = pd.DataFrame(d2)

print(df1)
print('-'*35)
print(df2)
```

### GroupBy

groupby will come up a lot of times whenever you want to aggregate your data. Pandas lets you do this efficiently with the groupby function.
There are a lot of ways that you can use groupby.

```python
# df.groupby(list of columns to groupby on).aggregate({'colname':func1, 'colname2':func2}).reset_index()
```

```python
test_df = df[['Status', 'Gender', 'Name', 'Sex', 'Age', 'Class']]

groups_df = test_df.groupby('Gender')
groups_df
```

```python
for group_col, group in groups_df:
    print("----- {} -----".format(group_col))
    print(group)
    print("")    
```

```python
for group_col, group in groups_df:
    print("----- {} -----".format(group_col
                                 ))
    print(group['Age'])
    print("")  
```

```python
for group_col, group in groups_df:
    print("----- {} -----".format(group_col))
    print('Mean Age: {:.2f}'.format(group['Age'].mean()))
    print("")  
```

```python
# SCHEMA: df.groupby().agg(col: func)

# 1. Groups the Data by Gender
# 2. Applies agg functions
# 3. Works only there the function can be applied

df.groupby('Gender').agg('mean')
```

```python
# Specify which columns to aggregate and how

# df.groupby('Gender').agg({'Age': 'mean', 'Class': 'median'})
print(df.groupby('Gender').agg({'Age': np.mean, 'Class': np.median})) # Groupby columns becomes the index of the dataframe!
print('-'*35)
print(df.groupby('Gender').agg({'Age': np.mean, 'Class': np.median}).reset_index()) # Allows to put the groupy by column as col!
```

```python
# Groupby by more than one column

df.groupby(['Gender', 'Embarked']).agg({'Age': np.mean, 'Encoded_Status': np.mean, 'ID': 'count'}).reset_index()
```

```python
# Now select Gender where the mean age is above its mean in the group

def own_mean(x):
    return np.mean(x)

mean_age_dict = df.groupby('Gender').agg({'Age': own_mean}).to_dict()['Age']
print(mean_age_dict)
print('-'*35)

print(df[['Age', 'Gender']].head(9))
print('-'*35)
print(df[df.apply(lambda x: x['Age'] > mean_age_dict[x['Gender']], axis=1)][['Age', 'Gender']].head())
```

### Melt

So in this melt function, we provide five attributes:
- dataframe_name = working_df
- id_vars: List of vars we want in the current form only.
- value_vars: List of vars we want to melt/put in the same column
- var_name: name of the column for value_vars
- value_name: name of the column for value of value_vars

```python
gender_unique = set(df.Gender)
    
for g in gender_unique:
    df[g] = df.apply(lambda x: 1 if x['Gender'] == g else 0, axis=1)
```

```python
df_to_melt = df[['ID', 'Sex', 'Name'] + list(gender_unique)]
df_to_melt.head()
```

```python
df_melted = pd.melt(df_to_melt, id_vars=['ID', 'Sex', 'Name'], var_name='Gender', value_name='Flag')
df_melted_sel = df_melted[df_melted['Flag']==1]
df_melted_sel.head()
```

### Pivot Table

We provide four attributes to the pivot_table function:
- index: We don’t want to change these column structures
- columns: explode this column into multiple columns
- values: use this column to aggregate
- aggfunc: the aggregation function.

```python
# Gender remains index column

pd.pivot_table(df_melted, index=['ID', 'Sex', 'Name'], columns='Gender', values='Flag', aggfunc=np.sum).reset_index().head()
```

```python
pd.pivot_table(df_melted_sel, index=['ID', 'Sex', 'Name'], columns='Gender', values='Flag', aggfunc=np.sum) \
    .reset_index() \
    .fillna(0) \
    .head()
```
